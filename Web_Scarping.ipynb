```python
import requests
from bs4 import BeautifulSoup
```

```python
l=['Artificial_neural_network','Support_vector_machine','Decision-making','Logistic_regression','Linear_regression']
```

```python
for i in l:
  url="https://en.wikipedia.org/wiki/"+str(i)
  response=requests.get(url)
```

```python
soup=BeautifulSoup(response.text,'html.parser')
```

```python
links=soup.find_all('a')
```

```python
for link in links:
  print(link.get('href'))
```
